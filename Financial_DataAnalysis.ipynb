## Data visualization

### Subtask:
Visualize the distribution of the data using histograms for numerical features and bar charts for categorical features.


**Reasoning**:
Visualize the distribution of numerical and categorical features using histograms and bar charts, respectively. Handle potential issues like data types and unique values before visualization.



```python
import matplotlib.pyplot as plt

# Numerical features visualization
numerical_features = ['age', 'Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 'Stock_Marktet', 'Factor']
for col in numerical_features:
    plt.figure(figsize=(8, 6))
    plt.hist(df[col], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Categorical features visualization
categorical_features = ['gender', 'Investment_Avenues', 'Duration', 'Invest_Monitor', 'Expect', 'Avenue', 'What are your savings objectives?', 'Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD', 'Source']
for col in categorical_features:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar', color='salmon')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
